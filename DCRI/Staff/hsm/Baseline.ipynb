```r
#Mean = function(x) base::mean(x, na.rm=TRUE)
#SD = function(x) stats::sd(x, na.rm=TRUE)

cont_mean = function(x){
    Meanx=round(mean(x, na.rm=TRUE), digits=1)
    SDx=round(sd(x, na.rm=TRUE), digits=1)
    val = paste(Meanx, " (", SDx, ")" , sep="")
    return(val)
}
cont_med = function(x){
    med=quantile(x, c(0.25, 0.5, 0.75), na.rm=TRUE)
    val=paste(med[2], "(", med[1],", ",med[3],")" , sep="")
    return(val)
}
cat_func = function(x){
    tab=CrossTable(all$study, x, na.rm=TRUE)
    n = tab$t
    prop=round((tab$prop.row)*100, digits=2)
    val = paste(n[,2], " (", prop[,2],"%)", sep="")
    val1= t(data.frame(study=rownames(n), label=val))
    return(val1)
}
cat_func(all$HxDM)

#tab=CrossTable(all$study, all$HxDM, na.rm=TRUE)
tabular(Heading("Systolic BP")*(BP.s)~ (study + 1)*Heading()*(cont_med), data=all)

```
