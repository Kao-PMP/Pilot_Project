```r
setwd('/mnt/workspace')
#getwd()
all=read.csv('combined-5.csv')
```

```r
library(dplyr)
```

```r
unique(all$study)
some=all %>% filter(study!='TOPCAT')
unique(some$study)
```

```r
colnames(some)
```

```r
final = some %>% arrange(person_id) %>%
group_by(person_id) %>% 
summarize(comp = max(mihosp_status, cva_status, cvdth_status), tcomp=min(cvdth_dt, mihosp_dt, cva_dt)) %>% 
right_join(some, by='person_id')
```

```r
head(final)
```

```r
final %>% group_by(study) %>% summarize(med_time = median(tcomp, na.rm=TRUE))
```

```r
head(final[final$study=='ALLHAT',c('person_id', 'BMI', 'weight','age','dth_dt','dth_status','cvdth_dt','mihosp_dt','cva_dt')])
```
